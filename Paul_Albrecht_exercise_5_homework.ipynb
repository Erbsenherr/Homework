# Task 7:> Look up `DESCRIBE` in the SPARQL Documentation and describe what it does:

The DESCRIBE form returns an RDF graph that describes the resources found. The exact set of triples returned for a resource is implementation-dependent (the SPARQL endpoint decides which properties to include), but typically includes a selection of triples where the resource is subject (and sometimes where it is object). DESCRIBE is useful for quickly getting RDF data about a resource without specifying a detailed CONSTRUCT template.
